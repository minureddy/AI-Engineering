```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hotel.drop_duplicates(inplace=True)
```

```python
df_hotel.duplicated().sum()
```

```python
df_hotel_dropna=df_hotel.drop(['company'],axis=1)
```

```python
df_hotel_dropna.isna().sum().sort_values(ascending=False)
```

```python
df_hotel_dropna['country'].value_counts()
```

```python
most_frequent_country=df_hotel_dropna['country'].value_counts().idxmax()
print(most_frequent_country)
```

```python
df_hotel_dropna['country'].fillna(most_frequent_country, inplace=True)
```

```python
df_hotel_dropna.isna().sum().sort_values(ascending=False)
```

```python
df_hotel_dropna['children']=df_hotel_dropna['children'].fillna(df_hotel_dropna['children'].mode()[0])
df_hotel_dropna['agent']=df_hotel_dropna['agent'].fillna(df_hotel_dropna['agent'].mode()[0])
```

```python
df_hotel_dropna.isna().sum().sort_values(ascending=False)
```

```python
df_hotel_dropna['country'].nunique()
```

```python
df_countryvshotel_counts = df_hotel_dropna.groupby(['hotel','country']).size().reset_index(name='booking_counts')
df_top_countryvshotel_counts=df_countryvshotel_counts.sort_values('booking_counts', ascending=False).groupby('hotel').head(10)
plt.figure(figsize=(12,5))
ax = sns.barplot(x='hotel', y='booking_counts', hue='country', data=df_top_countryvshotel_counts)

#ax.bar_label(ax.containers[0])
for container in ax.containers:
    ax.bar_label(container)
plt.title('Top Country Booking Most in Each Hotel')
plt.ylabel('Total Bookings')
plt.xlabel('Hotel')
plt.tight_layout()
plt.show()
```

```python
df_hotel_dropna['totalpeople']=df_hotel_dropna['adults']+df_hotel_dropna['babies'].fillna(0)+df_hotel_dropna['children'].fillna(0)
df_stayed=df_hotel_dropna[df_hotel_dropna['is_canceled'] == 0]
df_people_total=df_stayed.groupby('hotel')["totalpeople"].size().reset_index()

plt.figure(figsize=(6,4))
ax= sns.barplot(data=df_people_total,x='hotel',y='totalpeople')
for container in ax.containers:
    ax.bar_label(container)
```

```python
df_agg=df_hotel_dropna.groupby(["hotel", "is_canceled"]).size().reset_index(name='counts')
df_agg['percentage (%)']=100 * df_agg['counts']/df_agg.groupby("hotel") ['counts'].transform('sum')
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='hotel', y='percentage (%)', hue='is_canceled', data=df_agg)
for container in ax.containers:
  ax.bar_label(container,fmt='%1.1f%%')
plt.title('Cancellation Percentage by Hotel')
plt.ylabel('Percentage (%)')
plt.xlabel('Hotel Type')
plt.legend(title='Is Canceled')
plt.tight_layout()
plt.show()
```

```python
df_hotel_dropna['total_nights'] =df_hotel_dropna['stays_in_week_nights']+df_hotel_dropna['stays_in_weekend_nights']

df_stayed=df_hotel_dropna[df_hotel_dropna['is_canceled']==0]
df_avg_ppl= df_stayed.groupby('hotel')['total_nights'].mean().reset_index()
plt.figure(figsize=(8,8))
plt.pie(
    df_avg_ppl['total_nights'],
    labels=df_avg_ppl['hotel'],
    autopct='%1.1f%%',
    startangle=90,

)
plt.title('Average Nights Stayed in Each Hotel')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd

df_hotel_dropna['arrival_date'] = pd.to_datetime(
    df_hotel_dropna['arrival_date_year'].astype(str) + '-' +
    df_hotel_dropna['arrival_date_month'] + '-' +
    df_hotel_dropna['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d'#full month name
)

# Extract useful parts
df_hotel_dropna['arrival_year'] = df_hotel_dropna['arrival_date'].dt.year
df_hotel_dropna['arrival_month_num'] = df_hotel_dropna['arrival_date'].dt.month
df_hotel_dropna['arrival_month'] = df_hotel_dropna['arrival_date'].dt.strftime('%b')  # Jan,
df_hotel_dropna['arrival_dayofweek_num'] = df_hotel_dropna['arrival_date'].dt.dayofweek  # Monday=0
df_hotel_dropna['arrival_dayofweek'] = df_hotel_dropna['arrival_date'].dt.strftime('%a') # Mon, Tue...
```

```python
yearly_bookings= df_hotel_dropna.groupby(['arrival_year','hotel']).size().reset_index(name='count')


plt.figure(figsize=(16,8))
sns.lineplot(data=yearly_bookings,x='arrival_year',y='count',marker='o',hue='hotel')
plt.title('Yearly bookings per hotel')
plt.ylabel('Number of Bookings')
plt.xlabel('Year')
plt.grid(True)
plt.show()






```

```python
monthly_bookings = df_hotel_dropna.groupby(['arrival_year', 'arrival_month_num', 'hotel']).size().reset_index(name='count')
monthly_bookings = monthly_bookings.sort_values(['arrival_year', 'arrival_month_num'])
plt.figure(figsize=(14,7))
sns.lineplot(data=monthly_bookings, x='arrival_month_num', y='count', hue='hotel', style='arrival_year', markers=True)
plt.title('Monthly Bookings per Hotel (multiple years)')
plt.xlabel('Month (1=Jan)')
plt.ylabel('Number of Bookings')
plt.xticks(ticks=range(1,13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
weekday_bookings = df_hotel_dropna.groupby(['arrival_dayofweek_num', 'hotel']).size().reset_index(name='count')
weekday_bookings = weekday_bookings.sort_values('arrival_dayofweek_num')
plt.figure(figsize=(10,6))
sns.barplot(data=weekday_bookings, x='arrival_dayofweek_num', y='count', hue='hotel')
plt.title('Bookings by Day of Week per Hotel')
plt.xlabel('Day of Week (0=Mon)')
plt.ylabel('Number of Bookings')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python


yearly_combined = df_hotel_dropna.groupby('arrival_year').size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_combined, x='arrival_year', y='count', marker='o')
plt.title('Yearly Bookings (Both Hotels Combined)')
plt.xlabel('Year')
plt.ylabel('Number of Bookings')
plt.grid(True)
plt.show()
```

```python
monthly_combined = df_hotel_dropna.groupby(['arrival_year', 'arrival_month_num']).size().reset_index(name='count')
monthly_combined = monthly_combined.sort_values(['arrival_year', 'arrival_month_num'])

plt.figure(figsize=(14,7))
sns.lineplot(data=monthly_combined, x='arrival_month_num', y='count', hue='arrival_year', palette='tab10', marker='o')
plt.title('Monthly Bookings (Both Hotels Combined)')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(ticks=range(1,13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
weekly_combined = df_hotel_dropna.groupby(['arrival_year','arrival_dayofweek_num']).size().reset_index(name='count')
weekly_combined = weekly_combined.sort_values(['arrival_year','arrival_dayofweek_num'])

plt.figure(figsize=(10,6))
sns.barplot(data=weekly_combined, x='arrival_dayofweek_num', y='count', color='coral',hue='arrival_year')
plt.title('Bookings by Day of Week (Both Hotels Combined)')
plt.xlabel('Day of Week')
plt.ylabel('Number of Bookings')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
